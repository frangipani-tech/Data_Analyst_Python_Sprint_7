# **Nivel 1_Ejercicio 1: Selección de Personal**

El departamento de RRHH ha abierto una oferta laboral interna con un límite de
10 candidatos. El objetivo de este ejercicio es construir una herramienta que actúe como formulario de inscripción: recoge los datos de cada
candidato, calcula una puntuación según experiencia, conocimiento de catalán,
dominio de Excel y titulación académica, y los clasifica en tres niveles de
prioridad. Cuando se completan los 10 registros, genera un informe final
ordenado por prioridad y puntuación.




## Planteamiento

Voy a dividir el problema en diferentes funciones, cada una con una única
responsabilidad:

- obtener_datos: pide al usuario los datos del candidato mediante input
  y los devuelve en un diccionario;
- validar_datos: comprueba que los datos introducidos son correctos
  y convierte los strings a los tipos adecuados;
- calcular_puntuacion: aplica la fórmula experiencia * 1.5 + catalán +
  excel + título y devuelve la puntuación total;
- clasificar_candidato: a partir de la puntuación, devuelve la
  categoría de prioridad;
- mostrar_individual: imprime por pantalla el resultado de un
  candidato concreto tras inscribirse;
- mostrar_informe: ordena la lista por prioridad y puntuación e imprime el informe final;
- main_seleccion: gestiona todo el flujo. Repite el proceso hasta
  llegar a 10 candidatos y al final llama a mostrar_informe.

## Función obtener_datos

Esta función se encarga únicamente de recoger los datos del candidato a través de los input. Pide nombre y apellidos, teléfono, años de experiencia, y los niveles de catalán, Excel y titulación académica. Devuelve un diccionario con todos los campos, lo que facilita el acceso posterior por nombre en lugar de depender del orden de los valores.

```python
def obtener_datos():
    nombre_apellidos = input("Introduce tu nombre y tu apellido: ")
    telefono = input("Introduce tu teléfono de contacto: ")
    experiencia = input("Introduce tus años de experiencia dentro de la empresa: ")
    catalan = input("¿Cuál es tu conocimiento del Catalán? 0 = sin conocimiento; 1 = hasta nivel B1; 2 = nivel B2; 3 = C1 o superior: ")
    excel = input("¿Cuál es tu dominio de Excel? 0 = nivel básico; 1 = uso de funciones; 2 = tablas dinámicas; 3 = Visual Basic y Macros: ")
    titulo = input("¿Cuál es tu título académico? 0 = ESO; 1 = FP; 2 = Universidad; 3 = Máster o superior: ")
    return {
       "nombre": nombre_apellidos,
       "telefono": telefono,
       "experiencia": experiencia,
       "catalan": catalan,
       "excel": excel,
       "titulo": titulo,
   }
```

## Función validar_datos

Esta función se encarga de validar los datos del candidato. Comprueba que el
teléfono sea numérico, que los años de experiencia sean un número no negativo,
y que los valores de catalán, Excel y título estén entre 0 y 3. Además
convierte los strings al tipo adecuado (int o float). Si todo es correcto, devuelve un diccionario con los datos ya convertidos. Si algún dato está fuera de rango (por ejemplo, un catalán de 5), muestra un  mensaje de error específico y devuelve "None". Si algún dato no es numérico  y no se puede convertir (por ejemplo, "hola" en el campo de experiencia),  el bloque "except" captura el "ValueError", muestra un mensaje genérico y
también devuelve "None".

```python
def validar_datos(candidato):
    try:
        telefono_num = int(candidato["telefono"])
        experiencia_num = float(candidato["experiencia"])
        if experiencia_num < 0:
            print("Error: el numero de años de experiencia no puede ser negativo")
            return None
        catalan_num = int(candidato["catalan"])
        if catalan_num < 0 or catalan_num > 3:
            print("Error: El valor de catalán debe estar entre 0 y 3.")
            return None
        excel_num = int(candidato["excel"])
        if excel_num < 0 or excel_num > 3:
            print("Error: El valor de Excel debe estar entre 0 y 3.")
            return None
        titulo_num = int(candidato["titulo"])
        if titulo_num < 0 or titulo_num > 3:
            print("Error: El valor del título debe estar entre 0 y 3.")
            return None

        return {
            "nombre": candidato["nombre"],
            "telefono": telefono_num,
            "experiencia": experiencia_num,
            "catalan": catalan_num,
            "excel": excel_num,
            "titulo": titulo_num,
        }

    except ValueError:
        print("Error: Por favor, vuelve a intentarlo.")
        return None
```

## Función calcular_puntuación

Esta función se encarga únicamente de calcular la puntuación total de cada candidato. En otras palabras aplica la fórmula años de experiencia * 1.5 + catalán + excel + título y devuelve la puntuación total.

```python
def calculo_puntuacion(candidato):
    puntuacion = candidato["experiencia"] * 1.5 + candidato["catalan"] + candidato["excel"] + candidato["titulo"]
    return puntuacion
```

## Función clasificar_candidato

Esta función se encarga únicamente de clasificar cada candidato. A partir de
la puntuación recibida, devuelve la categoría de prioridad para RRHH según
los criterios establecidos:
- Puntuación >= 12 = "Prioridad Alta"
- Puntuación entre 8 y 11 = "Candidato adecuado"
- Puntuación < 8 = "No prioritario"

```python
def clasificar_candidato(puntuacion):
    if puntuacion >= 12:
        return "Prioridad Alta"
    elif puntuacion >= 8:
        return "Candidato adecuado"
    else:
        return "No prioritario"
```

## Funcion mostrar_individual

Esta función se encarga simplemente de imprimir por pantalla con el print el resultado de un candidato concreto tras inscribirse a la oferta de trabajo. Como su única responsabilidad es mostrar información, no devuelve ningún valor.

```python
def mostrar_individual(puntuacion):
    print(f"Has introducido correctamente todos los datos. Tu puntuación final es de {puntuacion}. ¡Muchas gracias por haberte apuntado a la oferta!")
```

## Función mostrar_informe

Esta función ordena la lista de candidatos e imprime el informe final.

Para el orden, el primer bloque "orden_prioridad" asigna un número a cada categoría de prioridad ("Prioridad Alta" = 1, "Candidato adecuado" = 2, "No prioritario" = 3) mediante un diccionario, ya que "sorted" no puede ordenar texto.

Luego tenemos el segundo bloque, lo del "sorted" que ordena la lista. El "key" dice: "para cada candidato, fíjate en estos valores para compararlo con los demás", es decir prioridad (del bloque anterior) y puntuación. El lambda devuelve para cada candidato (que llamo c) una tupla () de dos elementos: esto es importante porque sorted, al comparar tuplas, mira primero el primer elemento. Solo si hay empate, mira el segundo. Eso es justo lo que pide el enunciado. De esta forma, la lista se ordena primero por el número de prioridad y, en caso de empate, por puntuación descendente (de mayor a menor).

Merece la pena mencionar lo siguiente: la puntuación es en negativo porque sorted ordena de menor a mayor. Pero como queremos las puntuaciones altas primero, este truco invierte el orden.

Finalmente, recorre la lista ya ordenada e imprime para cada candidato su
nombre, teléfono y puntuación tal y como pide el enunciado, para que el departamento de RRHH pueda contactarles rápidamente.

```python
def mostrar_informe(candidatos):

    orden_prioridad = {
      "Prioridad Alta": 1,
      "Candidato adecuado": 2,
      "No prioritario": 3,
    }

    lista_ordenada = sorted(
        candidatos,                                                        # el qué ordenar, es decir los candidatos. La linea siguiente es cómo ordenar
        key=lambda c: (orden_prioridad[c["prioridad"]], -c["puntuacion"])  # Ordena por prioridad (criterio 1) y, en empate, por puntuación de mayor a menor (criterio 2, con el menos para invertir).
    )

    print("===== INFORME FINAL =====")
    for candidato in lista_ordenada:
        print(f"{candidato['nombre']} - {candidato['telefono']} - {candidato['puntuacion']}")
```

##Función main_seleccion

Esta función gestiona todo el flujo, básicamente va llamando a las demás funciones en orden. Repite el proceso hasta llegar a 10 candidatos y al final llama a mostrar_informe.

Empieza con una lista vacía, donde se van guardando los candidatos según se inscriban.

El while se repite mientras la lista tenga menos de 10 candidatos, y con el continue si un candidato mete datos inválidos no los incluye.

- datos = obtener_datos(): pide los datos al usuario.
- datos_validados = validar_datos(datos): le pasa esos datos a validar. Será el diccionario convertido (si todo bien) o None (si algo falló).
- if datos_validados is None: comprueba si la validación falló.
- continue: si falló, salta al inicio del while y vuelve a empezar.
- calculo = calculo_puntuacion(datos_validados): calcula la puntuación y la guarda en calculo.
clasificacion = clasificar_candidato(calculo): con esa puntuación, saca la categoría ("Prioridad Alta", etc.) y la guarda en clasificacion.
- mostrar_individual(calculo): imprime el mensaje individual al candidato con su puntuación.
- datos_validados["puntuacion"] = calculo:  crea la clave "puntuacion" dentro del diccionario, y le pone como valor lo que vale calculo (la puntuación que ya calculamos).
- datos_validados["prioridad"] = clasificacion: hace lo mismo, crea la clave "prioridad" y le pone el valor de clasificacion. Después de estas dos líneas, datos_validados ha pasado de tener 6 claves a tener 8. Es importante porque  mostrar_informe luego va a buscar candidato["puntuacion"] y candidato["prioridad"].
- candidatos.append(datos_validados): coge el diccionario datos_validados (el candidato ya completo, con sus 8 claves) y lo mete al final de la lista candidatos.
- mostrar_informe(candidatos): cuando el while termina (ya hay 10 candidatos), est línea se ejecuta una sola vez.


```python
def main_seleccion():
    candidatos = []

    while len(candidatos) < 10:
        datos = obtener_datos()
        datos_validados = validar_datos(datos)
        if datos_validados is None:
            continue
        calculo = calculo_puntuacion(datos_validados)
        clasificacion = clasificar_candidato(calculo)
        mostrar_individual(calculo)

        datos_validados["puntuacion"] = calculo
        datos_validados["prioridad"] = clasificacion
        candidatos.append(datos_validados)


    mostrar_informe(candidatos)
```

```python
main_seleccion()
```

La empresa quiere mejorar la efectividad de las campañas de recuperación de cestas de compra abandonados y necesita identificar qué clientes tienen mayor riesgo de abandono, qué segmento de comportamiento presentan y qué mensaje publicitario es el más adecuado para cada uno.



# Planteamiento

Voy a descomponer el problema en partes:

- primero voy a crear los diccionarios que las funciones consultarán.
- obtener_segmento_y_factor: recibe un código de comportamiento, es decir un número del 0 al 4. Busca ese código en el diccionario de comportamientos.
Devuelve el factor, el segmento y el mensaje de ese código.
- obtener_factor_edad: recibe el grupo de edad, busca ese grupo en el diccionario de factores de edad y devuelve el factor de edad correspondiente.
- calcular_riesgo: recibe el factor de comportamiento y el factor de edad, los multiplica y devuelve el riesgo final.
- procesar_clientes: recibe la lista de los 10 clientes. Para cada cliente llama a las tres funciones de arriba y junta toda la info. Devuelve la lista de clientes ya procesada y completa.

Como decía al principio, definiré 2 diccionarios que las funciones consultarán: uno para los comportamientos (código: factor, segmento, mensaje) y otro para los factores de edad (grupo: factor). Esto permite hacer la segmentación sin usar if/elif/else.

## Diccionarios de configuración

Antes de las funciones, defino los dos diccionarios que contienen los datos fijos  del ejercicio. Estos diccionarios son la pieza central de la solución ya que permiten hacer la segmentación sin usar if/elif/else.

"Comportamientos" es un diccionario de diccionarios. La clave es el código
de comportamiento, de 0 a 4, y el valor es otro diccionario con 3 datos asociados: el factor de riesgo, el segmento del cliente y el mensaje
publicitario (estos útltimos dos los defino yo). Los agrupo porque todos dependen del mismo código.

"Factores_de_edad" es un diccionario simple. La clave es el grupo de edad, como texto, y el valor es el factor de edad correspondiente.

De esta forma, en lugar de comprobar el código con condiciones, basta con acceder directamente.

» Códigos de comportamiento

0 → Navega pero no añade productos

1 → Agrega productos pero no llega al checkout

2 → Llega al checkout pero no paga

3 → Cesta de compra abandonado hace menos de 24 h

4 → Cesta de compra abandonado hace más de 24 h

» Factores según código de comportamiento

Código 0 → 0.10

Código 1 → 0.25

Código 2 → 0.50

Código 3 → 0.75

Código 4 → 0.90

» Factores según grupo de edad

18–25 → 1.20

26–40 → 1.00

41–60 → 0.85

60+ → 0.70

```python
comportamientos = {
    0: {"factor": 0.10, "segmento": "Visitante", "mensaje": "Descubre nuestros productos"},
    1: {"factor": 0.25, "segmento": "Explorador", "mensaje": "¿Aún te lo estás pensando?"},
    2: {"factor": 0.50, "segmento": "Comprador dudoso", "mensaje": "Termina tu compra"},
    3: {"factor": 0.75, "segmento": "Abandono caliente", "mensaje": "¡Tu cesta te espera!"},
    4: {"factor": 0.90, "segmento": "Abandono frío", "mensaje": "Vuelve y llévate un 10% de descuento"},
}

factores_de_edad = {
    "18-25": 1.20,
    "26-40": 1.00,
    "41-60": 0.85,
    "60+": 0.70,
}
```

##Función obtener_segmento_y_factor

Esta función recibe un código de comportamiento, es decir un número del 0 al 4, y  consulta el diccionario "comportamientos" para obtener los datos asociados
a ese código. Como "comportamientos" es un diccionario de diccionarios, al
acceder con la clave se obtiene el diccionario interior con el factor, el
segmento y el mensaje, que la función devuelve.

De esta forma la segmentación se hace consultando una clave del diccionario,
sin necesidad de usar if/elif/else.

- "info" guarda el diccionario interior del código recibido.
- "comportamientos[codigo]" devuelve el sub-diccionario con factor,
segmento y mensaje; después se accede a cada valor con "info["clave"]".

```python
def obtener_segmento_y_factor(codigo):                             # Define la función. Recibe un parámetro, codigo, que es el código de comportamiento de un cliente (un número del 0 al 4).
    info = comportamientos[codigo]                                 # Busca ese código en el diccionario comportamientos y guarda el resultado en la variable info
    return info["factor"], info["segmento"], info["mensaje"]       # saca los tres datos (factor, segmento y mensaje) y los devuelve.
```

## Función obtener_factor_edad

Esta función recibe un grupo de edad, es decir un texto, como "18-25", y consulta el diccionario factores_de_edad para obtener el factor correspondiente. Este es un diccionario simple, así que  al acceder con la clave se obtiene directamente el factor.

Igual que la función anterior, evita el uso de if/elif/else apoyándose
en la consulta directa al diccionario.

```python
def obtener_factor_edad(grupo_edad):
    return factores_de_edad[grupo_edad]
```

## Función calcular_riesgo

Esta función recibe el factor de comportamiento y el factor de edad, los multiplica y devuelve el riesgo final. factor_comportamiento y factor_edad son los parámetros de la función. Es decir: son los nombres de los dos números que la función recibe cuando la usaremos luego. Son etiquetas para los valores que entran.

```python
def calcular_riesgo(factor_comportamiento, factor_edad):
    riesgo_final = factor_comportamiento * factor_edad
    return riesgo_final
```

## Función procesar_clientes

Esta función se alimenta de las anteriores. En el detalle:
- Recibe la lista de los 10 clientes, y para cada cliente coge su codigo_comportamiento y se lo pasa a obtener_segmento_y_factor. Eso le devuelve 3 cosas: factor de comportamiento, segmento y mensaje.
- Coge su grupo_edad y se lo pasa a obtener_factor_edad. Eso le devuelve el factor de edad.
- Coge los dos factores (comportamiento y edad) y se los pasa a calcular_riesgo. Eso le devuelve el riesgo final.
- De esta forma tendremos 5 nuevos datos más por cliente, que añadiremos al diccionario de cada uno.

```python
def procesar_clientes(clientes):
    for cliente in clientes:
        factor_comportamiento, segmento, mensaje = obtener_segmento_y_factor(cliente["codigo_comportamiento"])
        factor_edad = obtener_factor_edad(cliente["grupo_edad"])
        riesgo_final = calcular_riesgo(factor_comportamiento, factor_edad)

        cliente["factor_comportamiento"] = factor_comportamiento    # Cada línea coge una variable y la mete dentro del diccionario cliente con su nombre de clave.
        cliente["segmento"] = segmento
        cliente["mensaje"] = mensaje
        cliente["factor_edad"] = factor_edad
        cliente["riesgo_final"] = riesgo_final

    return clientes
```

## Función mostrar_dataframe

Esta función recibe la lista de clientes procesada, luego la convierte a dataframe y la imprime de forma ordenada y legible. He añadido .round(2) para redondear decimales.

```python
def mostrar_dataframe(clientes):
    df = pd.DataFrame(clientes).round(2)
    display(df)
```

## Función mostrar_respuestas

Esta función recibe la lista de clientes procesada, analiza los datos para responder a las 4 preguntas de marketing e imprime las respuestas.

Las 4 preguntas son:
- ¿Qué clientes deben contactar primero en una campaña de retargeting?
- ¿Qué diferencias existen entre clientes con el mismo comportamiento pero de edades distintas? (Por ejemplo: ¿un cliente joven que abandona la cesta de compra hace 24 h tiene el mismo riesgo que uno de más edad?)
- ¿Qué clientes presentan tan poco interés que no vale la pena invertir presupuesto?
- ¿Qué cliente sería lo más rentable de recuperar?

Objetivo: Responder a las 4 preguntas de negocio del departamento de Marketing.


1. Pregunta 1 - Retargeting prioritario:
   - Ordenar clientes por riesgo_final (mayor a menor)
   - Mostrar top 5
2. Pregunta 2 - Comparación por edad
   - Agrupar clientes por código de comportamiento
   - Mostrar cada grupo con edad y riesgo de cada cliente
3. Pregunta 3 - Poco interés:
   - Ordenar clientes por riesgo_final (menor a mayor)
   - Mostrar bottom 5
4. Pregunta 4 - Más rentable:
   - Tomar el cliente con mayor riesgo_final

```python
def mostrar_respuestas(clientes):

    retargeting = sorted(clientes, key=lambda c: c['riesgo_final'], reverse=True)
    print("===== Retargeting =====")
    for cliente in retargeting[:5]:
        print(f"{cliente['nombre']} - {round(cliente['riesgo_final'], 2)} - {cliente['segmento']}")

    # sorted crea una nueva lista ordenada. necesita saber por qué criterio ordenar. El lambda se lo dice: "para cada cliente c, fíjate en su riesgo_final
    # La primera c (después de lambda) es el parámetro. Está diciendo: "esta mini-función recibe algo y lo voy a llamar c".
    # La segunda c (después de los :) usa ese parámetro. Está diciendo: "y lo que devuelvo es c["riesgo_final"]".
    # podría llamarse cliente para que se vea más claro.
    # Bucle sobre los 5 primeros de la lista ordenada

    print()
    clientes_por_comportamiento = {}
    for cliente in clientes:
        clientes_por_comportamiento.setdefault(cliente["codigo_comportamiento"], []).append(cliente)

    # Mete al cliente en la lista de su código de comportamiento; si esa lista aún no existe, créala vacía primero []
    # setdefault busca la clave en el diccionario. Si la clave existe, devuelve su valor (y no hace nada más). Si NO EXISTE → la crea con valor_por_defecto y lo devuelve

    print("===== Clientes con el mismo comportamiento pero de edades distintas =====")
    for codigo, lista_clientes in clientes_por_comportamiento.items():                    # .items(): en cada vuelta, da una pareja: clave y valor. codigo recibe la clave (0-4) y lista_clientes recibe la lista de clientes de ese código.
        print(f"Comportamiento {codigo}:")
        for cliente in lista_clientes:                                                    # Recorre la lista de ese grupo concreto e imprime cada cliente
            print(f"{cliente['nombre']} - {cliente['grupo_edad']} - riesgo: {round(cliente['riesgo_final'], 2)}")

    print()
    poco_interes = sorted(clientes, key=lambda c: c['riesgo_final'])
    print("===== Clientes de poco interes =====")
    for cliente in poco_interes[:5]:
        print(f"{cliente['nombre']} - {round(cliente['riesgo_final'], 2)} - {cliente['segmento']}")

    print()
    mas_rentable = retargeting[0]                                                         # Toma el primer elemento de retargeting, ya que que retargeting está ordenada de mayor a menor riesgo
    print("===== El cliente más rentable=====")
    print(f"{mas_rentable['nombre']} - {round(mas_rentable['riesgo_final'], 2)} - {mas_rentable['segmento']}")
```

## Función main_marketing

Define los datos de los 10 clientes, llama a procesar_clientes, luego a mostrar_diccionario y a mostrar_respuestas, en orden.

## Normalización del diccionario de niveles de riesgo

Creo una función que normaliza el diccionario de niveles de riesgo, reemplazando los guiones bajos por espacios para que coincida con el formato de los síntomas que vienen de los pacientes.

1. Creo un diccionario vacío
2. Para cada síntoma en el diccionario:
   - Reemplazo guiones bajos por espacios
   - Guardo el síntoma limpio con su valor numérico
3. Devuelve el diccionario normalizado

```python
def normalizar_diccionario_riesgo(nivel_riesgo):               # recibe como parámetro: nivel_riesgo, el diccionario

    diccionario_nivel_riesgo = {}
    for riesgo in nivel_riesgo:                                # Inicia un bucle que recorre cada llave
        sintoma_normalizado = riesgo.replace("_", " ")         # Limpiamos la llave insertando un espacio en lugar del guion bajo

        valor = nivel_riesgo[riesgo]                           # Obtiene el valor numérico asociado al síntoma original del diccionario

        diccionario_nivel_riesgo[sintoma_normalizado] = valor  # guardamos la nueva llave con su valor en el nuevo diccionario

    return diccionario_nivel_riesgo
```

## Procesamiento de los datos de pacientes

Creo una función que procesa y normaliza la lista de pacientes y convierte cada string en una estructura limpia, es decir un diccionario, normalizando los síntomas: todo en minúsculas, sin espacios extra, y reemplazando guiones bajos por espacios.

```python
def procesar_pacientes(pacientes):

    diccionario_pacientes = {}
    for paciente in pacientes:                                                         # recorre cada string de la lista pacientes
        primer_paso = paciente.lower().replace("|", ",").replace("_", " ")             # Transforma todo en minu, sostituye la | con la coma y remplaza guiones bajos por espacios

        paciente_limpio = [elemento.strip() for elemento in primer_paso.split(",")]    # Split divide el string por comas, creando una lista. Y elemento.strip(), para CADA elemento de esa lista, quita los espacios extra al principio y al final
                                                                                       # ejemplo paciente_limpio: ["maria", "45", "tos", "fiebre alta", "dolor pecho"]
        nombre = paciente_limpio[0]                                                    # Extrae el primer elemento de la lista (índice 0), que es el nombre del paciente
        edad = int(paciente_limpio[1])                                                 # Convierte la edad en num int, indice 1
        sintomas = paciente_limpio[2:]                                                 # Coje todas las sintomas, todos los elementos desde el índice 2 hasta el final


        diccionario_pacientes[nombre] = {                                              # Crea la estructura del diccionario: la llave es el nombre del pacientes
            "edad": edad,                                                              # Valor: otro diccionario con 2 campos: "edad" en número entero y "sintomas", la lista de síntomas
            "sintomas": sintomas
        }

    return diccionario_pacientes
```

## Cálculo del riesgo total por paciente

Creo una función que calcula el riesgo total de un paciente sumando los valores de riesgo de cada uno de sus síntomas según el diccionario normalizado de niveles de riesgo.

```python
def calcular_riesgo_paciente(sintomas, diccionario_nivel_riesgo):     # recibe dos parametros: la lista de sintoma de un paciente + el diccionario con los valores de riesgo de cada sintoma
    puntuacion_total = 0                                              # Inicializa una variable puntuacion_total en 0

    for sintoma in sintomas:                                          # Inicia un bucle que recorre cada síntoma de la lista del paciente
        if sintoma in diccionario_nivel_riesgo:                       # Comprueba si el síntoma actual existe como llave en el diccionario de niveles de riesgo
            puntuacion_total += diccionario_nivel_riesgo[sintoma]     # diccionario_nivel_riesgo[sintoma] Busca el síntoma actual en el diccionario y obtiene su valor numérico. y lo suma

    return puntuacion_total
```

## Añadir riesgo total a cada paciente

Creo una función que recorre todos los pacientes, calcula su riesgo total usando la función anterior, y lo añade a su información en el diccionario.

```python
def añadir_riesgo_total(diccionario_pacientes, diccionario_nivel_riesgo):
    for nombre, datos in diccionario_pacientes.items():                          # Recorre el diccionario de pacientes. en cada vuelta obtiene nombre y los valores asociados. El método .items() permite acceder a la llave y al valor al mismo tiempo.
        sintomas = datos["sintomas"]                                             # Extrae la lista de síntomas del diccionario datos y la guarda en la variable sintomas. # datos es un diccionario que contiene toda la información de UN paciente.

        riesgo = calcular_riesgo_paciente(sintomas, diccionario_nivel_riesgo)    # Calcula el riesgo usando la función anterior

        datos["riesgo_total"] = riesgo                                           # Guardar el riesgo en el diccionario del paciente. Añade un nuevo campo llamado "riesgo_total"

    return diccionario_pacientes                                                 # Devuelve el diccionario completo de pacientes, ahora actualizado con el campo "riesgo_total" en cada uno
```

## Ordenar pacientes por riesgo

Creo una función que ordena los pacientes de mayor a menor riesgo total para determinar la prioridad de atención.

```python
def ordenar_pacientes_por_riesgo(diccionario_pacientes):
    pacientes_ordenados = sorted(
        diccionario_pacientes.items(),          # .items() convierte el diccionario en una lista de tuplas
        key=lambda x: x[1]["riesgo_total"],     # Define por qué criterio ordenar | lambda x: función anónima que recibe x (cada tupla) | x[1]: accede al 2o elemento de la tupla | x[0] = nombre, x[1] = datos ordena por el valor de "riesgo_total" de cada paciente
        reverse=True
    )
    return dict(pacientes_ordenados)             # Convierte la lista de tuplas ordenada de nuevo en un diccionario
```

## Crear DataFrame final

Creo una función que convierte el diccionario de pacientes ordenado en un DataFrame de pandas con las columnas: nombre, edad, sintomas y riesgo_total.

```python
def crear_dataframe_pacientes(pacientes_ordenados):
    lista_datos = []

    for nombre, datos in pacientes_ordenados.items():
        lista_datos.append({
            "nombre": nombre,
            "edad": datos["edad"],
            "sintomas": ", ".join(datos["sintomas"]),             # .join() une elementos de una lista en un solo string. ", " es el separador que se pone entre cada elemento
            "riesgo_total": datos["riesgo_total"]
        })

    df = pd.DataFrame(lista_datos)
    return df
```

## Función principal

Creo la función principal que orquesta todo el proceso: normaliza el diccionario de riesgo, procesa los pacientes, calcula y añade el riesgo total, ordena por prioridad y crea el DataFrame final.

```python
def main_gestion_pacientes():

    diccionario_nivel_riesgo = normalizar_diccionario_riesgo(nivel_riesgo)                         # Normalizar el diccionario de riesgo
    diccionario_pacientes = procesar_pacientes(pacientes)                                          # Procesar los pacientes
    diccionario_pacientes = añadir_riesgo_total(diccionario_pacientes, diccionario_nivel_riesgo)   # Añadir el riesgo total a cada paciente
    diccionario_pacientes_ordenado = ordenar_pacientes_por_riesgo(diccionario_pacientes)           # Ordenar los pacientes por riesgo

    df_final = crear_dataframe_pacientes(diccionario_pacientes_ordenado)                           # Crear el DataFrame final

    print("DataFrame final con pacientes ordenados por riesgo:")
    print(df_final)

    return df_final
```

Nota técnica: En Python, True vale 1 y False vale 0, por lo que sumar booleanos cuenta cuántos están activos.

```python
def validar_parametros(longitud, majusculas, minusculas, numeros, signos):               # Valida que los parámetros sean correctos antes de generar la contraseña.

    if longitud <= 0:                                                                    # Validación 1: Longitud debe ser mayor que 0
        raise ValueError("Error: La longitud debe ser mayor que 0")

    if not (majusculas or minusculas or numeros or signos):                              # Validación 2: Al menos un tipo debe estar activo. El or verifica que al menos uno sea True.
        raise ValueError("Error: Debes seleccionar al menos un tipo de carácter")

    cantidad_tipos = majusculas + minusculas + numeros + signos                          # Validación 3: Longitud suficiente para todos los tipos
    if longitud < cantidad_tipos:
        raise ValueError(f"Error: Longitud mínima debe ser {cantidad_tipos}")

# sin RETURN: No necesita devolver nada, solo verificar

# Apuntes sobre Validación 3: Longitud suficiente
# En Python: True = 1, False = 0
# Esto permite sumar booleanos para contar cuántos están activos

# Ejemplo:
# True + True + False + True = 1 + 1 + 0 + 1 = 3

# Aplicado al generador de contraseñas:
# Si majuscules=True, minuscules=False, numeros=True, signes=True
# Entonces: cantidad_tipos = 1 + 0 + 1 + 1 = 3
# Necesito longitud >= 3 para meter al menos 1 carácter de cada tipo activo
```

## Función construir_conjunto_caracteres

La función crea un string con todos los caracteres disponibles según los tipos seleccionados. Dicho en otras palabras, define los caracteres disponibles que se pueden usar para construir la contraseña.

- Inicializa un string vacío
- Va añadiendo caracteres según cada tipo esté activo (True):
   - Mayúsculas: A-Z
   - Minúsculas: a-z
   - Números: 0-9
   - Signos: !$%&/()?¿|#@[]_-
- Devuelve el string completo con todos los caracteres disponibles

Nota técnica: Usamos += para concatenar (añadir) strings, no = que sobreescribiría el contenido anterior.

```python
def construir_conjunto_caracteres(mayusculas, minusculas, numeros, signos):
    caracteres_disponibles = ""                                                     # creo un string vacío donde iremos acumulando todos los caracteres posibles

    if mayusculas:
        caracteres_disponibles += "ABCDEFGHIJKLMNOPQRSTUVWXYZ"                      # El += concatena (añade al final)

    if minusculas:
        caracteres_disponibles += "abcdefghijklmnopqrstuvwxyz"

    if numeros:
        caracteres_disponibles += "0123456789"                                      # pongo los numeros como string, no podemos concatenar un número a un string

    if signos:
        caracteres_disponibles += "!$%&/()?¿|#@[]_-"

    return caracteres_disponibles
```

## Función garantizar_uno_de_cada:

Esta función asegura que la contraseña incluya al menos un carácter de cada tipo seleccionado.

Funcionamiento:
- Crea una lista vacía para guardar los caracteres garantizados
- Por cada tipo activo (True):
   - Elige 1 carácter aleatorio de ese tipo usando np.random.choice()
   - Lo añade a la lista con .append()
- Devuelve la lista con los caracteres garantizados

Ejemplo: Si mayusculas=True, minusculas=False, numeros=True, signos=False:
- Elige 1 mayúscula aleatoria: 'K'
- Elige 1 número aleatorio: '7'
- Devuelve: ['K', '7']

Nota: Estos caracteres garantizan que la contraseña cumple los requisitos mínimos. El resto se completará después de forma aleatoria.

```python
def garantizar_uno_de_cada(mayusculas, minusculas, numeros, signos):
    caracteres_garantizados = []

    if mayusculas:
        mayuscula_aleatoria = np.random.choice(list("ABCDEFGHIJKLMNOPQRSTUVWXYZ"))
        caracteres_garantizados.append(mayuscula_aleatoria)

    if minusculas:
        minuscula_aleatoria = np.random.choice(list("abcdefghijklmnopqrstuvwxyz"))
        caracteres_garantizados.append(minuscula_aleatoria)

    if numeros:
        numero_aleatorio = np.random.choice(list("0123456789"))
        caracteres_garantizados.append(numero_aleatorio)

    if signos:
        signo_aleatorio = np.random.choice(list("!$%&/()?¿|#@[]_-"))
        caracteres_garantizados.append(signo_aleatorio)

    return caracteres_garantizados
```

## Función longitud_deseada

Esta función sirve para completar hasta la longitud deseada, es decir rellenar el resto de la contraseña con caracteres aleatorios del conjunto disponible.

Funcionamiento:
- Calcula cuántos caracteres faltan: longitud - len(caracteres_garantizados)
- Elige esa cantidad de caracteres aleatorios del conjunto completo usando np.random.choice() con el parámetro `size`
- Añade esos caracteres aleatorios a la lista de garantizados usando +=
- Devuelve la lista completa (garantizados + aleatorios)

Ejemplo:
- Longitud deseada: 12
- Caracteres garantizados: ['K', '7'] (2 caracteres)
- Caracteres faltantes: 12 - 2 = 10
- Elige 10 caracteres aleatorios y los añade
- Devuelve lista con 12 caracteres en total

Nota: Uso +=  en lugar de .append() para añadir elementos individuales, y no el array completo.

```python
def longitud_deseada(longitud, caracteres_garantizados, caracteres_disponibles):
    caracteres_faltantes = longitud - len(caracteres_garantizados)
    caracteres_aleatorios = np.random.choice(list(caracteres_disponibles), size=caracteres_faltantes)
    caracteres_garantizados += list(caracteres_aleatorios)

    return caracteres_garantizados
```

## Función mezclar

Esta función sirve para mezclar aleatoriamente, es decir desordenar todos los caracteres para que la contraseña final no sea predecible.

```python
def mezclar(todos_caracteres):
    np.random.shuffle(todos_caracteres)
    return todos_caracteres

# el parámetro se puede llamar caracteres_completos, porque en este punto del flujo ya contiene TODOS los caracteres (garantizados + aleatorios)
```

## Función main_generador_password

Hemos llegado a la función principal, es decir la que gestiona todas las funciones anteriores, recibe los parámetros (longitud, majuscules, minuscules, numeros, signes) y devuelve la contraseña generada.

Parámetros, tal y como pide el enunciado:
- longitud: Número de caracteres de la contraseña
- mayusculas: Incluir letras mayúsculas (por defecto True)
- minusculas: Incluir letras minúsculas (por defecto True)
- numeros: Incluir números (por defecto True)
- signos: Incluir caracteres especiales (por defecto False)

Funcionamiento paso a paso:
1. Valida los parámetros: Verifica que la longitud sea válida y que haya al menos un tipo de carácter seleccionado
2. Construye el conjunto disponible: Crea el string con todos los caracteres posibles según los tipos activos
3. Garantiza los tipos: Selecciona al menos 1 carácter de cada tipo activo (obligatorio)
4. Completa la longitud: Añade caracteres aleatorios hasta alcanzar la longitud deseada
5. Mezcla y Desordena todos los caracteres para que la contraseña sea impredecible
6. Convierte a string: Une la lista de caracteres en un string usando ''.join()
7. Copia al portapapeles: Intenta copiar la contraseña (no funciona en Google Colab)
8. Devuelve la contraseña generada como string.

```python
def main_generador_password(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):
    validar_parametros(longitud, mayusculas, minusculas, numeros, signos)                                 # validar parámetros
    caracteres_disponibles = construir_conjunto_caracteres(mayusculas, minusculas, numeros, signos)       # Construye el conjunto de caracteres disponibles
    caracteres_garantizados = garantizar_uno_de_cada(mayusculas, minusculas, numeros, signos)             # garantiza al menos uno de cada tipo
    todos_caracteres = longitud_deseada(longitud, caracteres_garantizados, caracteres_disponibles)        # completa hasta la longitud deseada
    caracteres_mezclados = mezclar(todos_caracteres)                                                      # mezcla aleatoriamente
    contrasena = ''.join(caracteres_mezclados)                                                            # convierte la lista a string. .join() junta todos los elementos de una lista en un string, sin separadores.
    copiar_portapapeles(contrasena)                                                                       # sirve para copiar al portapapeles
    return contrasena
```

## Función lectura_archivo

Esta función lee el archivo de partidos y devuelve una lista de diccionarios, donde cada diccionario representa un partido.

Recibe como parámetro un nombre generico para hacerla flexible y reutilizable.

Funcionamiento línea por línea:
- Abre el archivo en modo lectura con codificación UTF-8 (para caracteres especiales como ñ, à)
- Crea una lista vacía para guardar todos los partidos
- Recorre el archivo línea por línea
- Limpia cada línea eliminando espacios y saltos de línea al inicio y final con strip()
- Separa la línea en 3 partes usando TAB como separador: equipo local, resultado, equipo visitante
- Divide el resultado por el guion para obtener los goles de cada equipo
- Convierte los goles de string a enteros
- Crea un diccionario con toda la información del partido
- Añade el diccionario a la lista de partidos
- Devuelve la lista completa con todos los partidos procesados

Resultado: lista con 500 diccionarios, uno por cada partido.

```python
def lectura_archivo(nombre_archivo):                                   # función que recibe parametro con nombre generico, más flexible y reutilizable, para que se pueda usar por ejemplo con otros archivos
    with open(nombre_archivo, 'r', encoding='utf-8') as archivo:       # abre el archivo en modo lectura, con codificación UTF-8 (para caracteres especiales como ñ, à). El with cierra el archivo automáticamente al terminar.
        listado_partidos = []                                          # Crea una lista vacía donde se guardarán todos los partidos
        for linea in archivo:                                          #  # aqui se procesa cada linea
            linea = linea.strip()                                      # Elimina espacios en blanco y saltos de línea (\n) al inicio y final de la línea.
            partes = linea.split('\t')                                 # Divide la línea en 3 partes usando TAB como separador. Ejemplo: ["Manlleu", "0-1", "Granollers"]

            equipo_local = partes[0]                                   # Extrae el primer elemento de la lista: el nombre del equipo local
            resultado = partes[1]                                      # Extrae el segundo elemento: el resultado del partido
            equipo_visitante = partes[2]                               # Extrae el tercer elemento: el nombre del equipo visitante

            goles = resultado.split('-')                               # Divide el resultado por el guion

            goles_local = int(goles[0])                                # Convierte el primer número (string) a entero. Es el número de goles del equipo local.
            goles_visitante = int(goles[1])                            # Convierte el segundo número (string) a entero. Es el número de goles del equipo visitante.

            partido = {                                                # Crea un diccionario con toda la información del partido
            "local": equipo_local,
            "goles_local": goles_local,
            "goles_visitante": goles_visitante,
            "visitante": equipo_visitante
            }

            listado_partidos.append(partido)                          # Añade el diccionario del partido a la lista de todos los partidos

    return listado_partidos                                           # Devuelve la lista completa con los partidos procesados
```

## Función procesar_partidos

Esta función recorre todos los partidos y construye un diccionario con los datos calculados de cada equipo:
   - Goles marcados
   - Goles recibidos
   - Puntos (Victoria = 3 pts, Empate = 1pt, Derrota = 0 pts)
   - Victorias, empates y derrotas

Recibe la lista de partidos devuelta por lectura_archivo().

Funcionamiento:
- Crea un diccionario vacío para guardar las estadísticas de todos los equipos
- Recorre cada partido de la lista
- Para cada equipo (local y visitante), verifica si ya existe en el diccionario
- Si no existe, lo crea con todos los valores inicializados en 0: goles marcados, goles recibidos, victorias, empates, derrotas y puntos
- Actualiza los goles marcados y recibidos de ambos equipos sumando los del partido actual
- Compara los goles para determinar el resultado:
  - Si goles local > goles visitante: local suma 1 victoria y 3 puntos, visitante suma 1 derrota
  - Si goles visitante > goles local: visitante suma 1 victoria y 3 puntos, local suma 1 derrota
  - Si son iguales: ambos suman 1 empate y 1 punto cada uno
- Devuelve el diccionario completo con las estadísticas de todos los equipos

Resultado: diccionario con estadísticas acumuladas de cada equipo después de procesar los 500 partidos.

```python
def procesar_partidos(listado_partidos):
    estadisticas_equipos = {}

    for partido in listado_partidos:
        if partido["local"] not in estadisticas_equipos:             # Verifica si el equipo local ya existe en el diccionario de estadísticas. Si no existe, lo crea con todos sus contadores en 0.
            estadisticas_equipos[partido["local"]] = {
            "goles_marcados": 0,
            "goles_recibidos": 0,
            "victorias": 0,
            "empates": 0,
            "derrotas": 0,
            "puntos": 0,
             }
        if partido["visitante"] not in estadisticas_equipos:         # Verifica si el equipo visitante ya existe en el diccionario de estadísticas. Si no existe, lo crea con todos sus contadores en 0.
            estadisticas_equipos[partido["visitante"]] = {
            "goles_marcados": 0,
            "goles_recibidos": 0,
            "victorias": 0,
            "empates": 0,
            "derrotas": 0,
            "puntos": 0,
             }
        estadisticas_equipos[partido["local"]]["goles_marcados"] += partido["goles_local"]            # Suma los goles que marcó el equipo local en este partido a su total acumulado.
        estadisticas_equipos[partido["local"]]["goles_recibidos"] += partido["goles_visitante"]       # Suma los goles que recibió el equipo local (los que marcó el visitante) a su total acumulado.
        estadisticas_equipos[partido["visitante"]]["goles_marcados"] += partido["goles_visitante"]    # Suma los goles que marcó el equipo visitante a su total.
        estadisticas_equipos[partido["visitante"]]["goles_recibidos"] += partido["goles_local"]       # Suma los goles que recibió el visitante (los que marcó el local) a su total.

        if partido["goles_local"] > partido["goles_visitante"]:                   # Compara: ¿ganó el equipo local? Si ganó el local: suma 1 victoria y 3 puntos.
            estadisticas_equipos[partido["local"]]["victorias"] += 1
            estadisticas_equipos[partido["local"]]["puntos"] += 3
            estadisticas_equipos[partido["visitante"]]["derrotas"] += 1           # El visitante perdió: suma 1 derrota (0 puntos).

        elif partido["goles_visitante"] > partido["goles_local"]:                 # Compara: ¿ganó el equipo visitante? Visitante gana: suma 1 victoria y 3 puntos.
            estadisticas_equipos[partido["visitante"]]["victorias"] += 1
            estadisticas_equipos[partido["visitante"]]["puntos"] += 3
            estadisticas_equipos[partido["local"]]["derrotas"] += 1               # Local pierde: suma 1 derrota.

        else:                                                                     # Si no ganó ninguno, fue empate. Ambos equipos suman 1 empate y 1 punto cada uno.
            estadisticas_equipos[partido["visitante"]]["empates"] += 1
            estadisticas_equipos[partido["local"]]["empates"] += 1
            estadisticas_equipos[partido["visitante"]]["puntos"] += 1
            estadisticas_equipos[partido["local"]]["puntos"] += 1

    return estadisticas_equipos
```

## Función equipo_mas_goleador

Sirve para encontrar el equipo más goleador: Busca el equipo con más goles marcados en el diccionario de estadísticas.

Recibe el diccionario de estadísticas devuelto por procesar_partidos().

Funcionamiento:
- Usa la función max() para buscar el máximo entre los nombres de los equipos
- El parámetro key indica cómo comparar: usando el valor de goles_marcados de cada equipo
- lambda crea una función temporal que extrae los goles_marcados de cada equipo
- max() compara todos esos valores y devuelve el nombre del equipo con el número más alto

Devuelve: el nombre del equipo más goleador (string).

```python
def equipo_mas_goleador(estadisticas_equipos):
    equipo_mas_goleador = max(estadisticas_equipos, key=lambda equipo: estadisticas_equipos[equipo]["goles_marcados"])

    return equipo_mas_goleador

# max(estadisticas_equipos... Busca el máximo entre las claves del diccionario (los nombres de los equipos)
# key le dice a max() cómo comparar los equipos
# lambda equipo crea una función temporal que recibe cada nombre de equipo

# estadisticas_equipos[equipo]["goles_marcados"]: Para cada equipo, extrae su valor de "goles_marcados"
# max() compara esos valores y devuelve el equipo con el número más alto
```

## Función equipo_mas_goleado

Sirve para encontrar equipo más goleado: Busca el equipo con más goles recibidos en el diccionario de estadísticas. Basicamente igual que la anterior, pero extraemos los goles recibidos.

```python
def equipo_mas_goleado(estadisticas_equipos):
    equipo_mas_goleado = max(estadisticas_equipos, key=lambda equipo: estadisticas_equipos[equipo]["goles_recibidos"])

    return equipo_mas_goleado
```

## Función clasificacion_ordenada

Sirve para crear la clasificación ordenada, es decir que sirve para ordenar los equipos por puntos de mayor a menor.

Recibe el diccionario de estadísticas devuelto por procesar_partidos().

Funcionamiento:
- Usa la función sorted() para ordenar los nombres de los equipos
- El parámetro key indica cómo ordenar: usando el valor de puntos de cada equipo
- El parámetro reverse=True invierte el orden para que sea de mayor a menor (descendente)
- lambda extrae los puntos de cada equipo para la comparación

Devuelve: lista con los nombres de los equipos ordenados por puntos (el primero es el líder).


```python
def clasificacion_ordenada(estadisticas_equipos):
    clasificacion_ordenada = sorted(estadisticas_equipos, key=lambda equipo: estadisticas_equipos[equipo]["puntos"],reverse=True)

    return clasificacion_ordenada
```
